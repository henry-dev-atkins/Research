```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
import openpyxl 
import csv
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime
%matplotlib inline
```

```python
def plot_two(data1, data2, t, size=(10,5)):

    fig, ax1 = plt.subplots(figsize=size)

    color = 'tab:red'
    ax1.set_xlabel('Month')
    ax1.set_ylabel(str(data1.name), color=color)
    ax1.plot(t, data1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel(str(data2.name), color=color)  # we already handled the x-label with ax1
    ax2.plot(t, data2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
    return
```

```python
start = datetime.datetime(2004, 1, 1)
#end = datetime.datetime(2021, 1, 1)

gold =  web.get_data_yahoo('GC=F','01/01/2004', '06/30/2021',interval='d')['Adj Close']

gold = pd.DataFrame(gold)
gold.index = pd.to_datetime(gold.index)
gold = gold.resample('1M').mean()

google_data = pd.read_csv('Inflation 2004.csv')


new_index = google_data['Month']
google_data.set_index(new_index, inplace=True)
google_data['Gold'] = gold.values
plot_two(google_data['Inflation rate: (Worldwide)'].rolling(7).mean(), google_data['Gold'].rolling(7).mean(), 
         t = google_data['Month'], size=(15,5))
```
